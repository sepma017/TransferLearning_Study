# 기울기 기반 최적화
- 최적화는 기본적으로 어떤 함수 $f(x)$를 최소화하거나 최대화하는 작업
- $x$는 숫자 백터 또는 스칼라
- $f(x)$는 **목적 함수(objective function)** 또는 **척도(criterion)**라고 부른다.(신경망에서는 비용함수, 손실함수, 오차함수라고 부름)
- 최소화하고자 하는 손실 함수는 $E$

- $y=f(x)$가 있다고 가정하면 $x$와 $y$는 실수.
- 이 함수의 미분은 $x$에 대한 작은 변화가 이 함수를 어떻게 변화시키는지 알려준다.
- 따라서 미분한 $x$를 극단적으로 변화시킴으로써 함수의 값을 줄이는 데 사용할 수 있다.
- $x$에서 $f^\prime(x)>0$이라고 가정하자. => 양의 방향으로 $x$를 증가시키면 $f(x)$가 증가함. 결국 충분히 작은 $ϵ$에 대해서 $f(x-ϵ)<f(x)$가 됨을 의미
- $f(x)$는 도함수의 '반대'방향으로 작은 보폭으로 $x$를 움직여서 줄일 수 있다.

전이학습_2-1.png

- $f^\prime(x)=0$이면 이 도함수는 최소 함수에 도달하기 위해 이동해야 하는 방향에 대한 정보를 제공하지 않는다.
- 미분은 국소 최적(최소/최대)에서 0일 수 있다.
- $x^*$에서 함수의 값 $f(x)$가 $x^*$의 모든 인접점보다 작은 경우 점 $x^*$를 **국소 최솟값(local minimum)**이라고 한다.
- 국소 최댓값도 정의할 수 있다.
- 어떤 점은 최댓값도 최솟값도 될 수 없는데, 그 점에서 미분 $f^\prime(x)$는 0 => **안장점(saddle point)**
- $f^\prime(x)=0$인 세 가지 시나리오

전이학습_2-2.png

- 가능한 모든 $x$값 중에서 $f$값이 가장 낮은 점을 **전역 최솟값(global minimum)**이라고 한다.
- 함수에는 하나 또는 여러 개의 전역 최솟값이 있을 수 있다.
- 그런데 전역 최솟값이 아닌 국소 최솟값이 있을 수도 있다.
- 반면 함수가 블록이면 전역 최솟값이 하나만 있고 국소 최솟값은 없음을 보장한다.

- 머신러닝에서는 몇 가지 변수 $f:R^n → R$의 실숫값 함수를 최소화하는데 관심이 있다.
- 여러 변수의 실숫값 함수의 간단한 예는 $\vec{x} = (x_1, x_2)$의 열판 온도 함수(hotplate temperature function)인 $f(x_1, x_2) = 50-x^2_1-2x^2_2$이다.
- 딥러닝에서는 일반적으로 손실 함수를 최소화하려고 하는데, 최적화를 어렵게 만드는 경우들이 있다.
    - 1) 신경망에서 가중치와 같은 여러 변수의 함수들은 국소 최솟값을 가지는 경우가 자주 있다.
    - 2) 많은 안장점이 매우 평평한 지역으로 둘러싸여 있다.
    - 3) 전역 최솟값이 있을 때도 있고 없을 때도 있다.
- 여러 변수를 가진 함수의 미분은 편도함수로 표현되는데, 나머지 미분 변수는 일정하게 유지하면서 입력 변수 $x_i$중 하나를 변경해서 함수의 변화율을 측정한다.
- 모든 변수에 관한 부분 미분의 벡터를 $f$의 **기울기 벡터(gradient vector)**라고 하며 $∇f$로 표시한다.
- 기울기 벡터 $∇f$를 벡터의 방향으로 투영해서 계산한다.
- 유닛 벡터 $\vec{v}$의 방향 내적은 $∇f⋅\vec{v}$이다.($\vec{v}$의 방향에서 $f$의 **방향 미분 계수(directional derivative)**라고 불리며, 종종 $∇_{\vec{v}}$로 표시한다.
- $f$를 최소화하기 위해서는 $f$값이 최대로 줄어드는 $\vec{x}$를 변화시키는 방향 $\vec{u}$를 찾아야 한다.
- $\vec{x}_a$가 $\vec{x}$와 매우 가까운 점이라면 $||\vec{x}-\vec{x}_a||$는 매우 작다.
- $\vec{x}$에 대한 테일러 급수의 확장은 다음과 같다.

- 방정식의 마지막 항이 $\vec{x}$에 충분히 가까워지려면 $\vec{x}_a$가 충분히 작아야 한다.
- 두 번째항은 $\vec{x}_a-\vec{x}$에 따른 $f$의 방향 미분을 나타낸다.

- $f(\vec{x})$는 $cos(θ)$가 $θ=π$일 때 최솟값인 -1에서 최대로 축소된다.
- $\vec{x}_a-\vec{x}$가 기울기 벡터 $∇f$와 반대 방향을 가리켜야 한다.
- 이것이 가장 가파른 하강 방향 $-∇f$ 또는 **가장 가파른 경사 하강(steepest gradient descent)** 방향이다.

- 열판(Hot plate) 예제: 주어진 좌표에서의 온도는 $f(x,y)=50-y^2-2x^2$의 함수로 주어짐.
    - 온도가 50인 중심(0,0)이 가장 뜨거운 지점.
    - 점 (x, y)에서의 그레디언트 벡터는 $∇f=(-4x, -2y)$로 주어짐.
    - 열판의 점 (2.3, 2)에서의 온도는 40 (이 점은 일정한 온도의 등고선에 있다.)
    - 보폭 ϵ이 기울기 반대 방향으로 이동하면 온도가 30도 줄어듦.(빨간색 화살표)

```python
## 텐서플로를 이용해 열판 온도 함수에 대한 경사 하강 최적화 구현하기

import tensorflow as tf
# x, y = (2, 2) 경사 하강 초기화
x = tf.Variable(2, name='x', dtype=tf.float32)
y = tf.Variable(2, name='y', dtype=tf.float32)

temperature = 50 - tf.square(y) - 2*tf.square(x)

# Gradient Descent Optimizer 초기화
optimizer = tf.train.GradientDescentOptimizer(0.1)  # 0.1은 학습율이다.
train = optimizer.minimize(temperature)
grad = tf.gradients(temperature, [x, y])    # 기울기 벡터를 계산해 보자.

init = tf.global_variables_initializer()
with tf.Session() as session:
    session.run(init)
    print("Starting at coordinate x={}, y={} and temperature there is. {}".format(session.run(x), session.run(y), session.run(temperature)))
    grad_norms=[]
    for step in range(10):
        session.run(train)
        g = session.run(grad)
        print("step ({}) x={}, y={}, T={}, Gradient={}".format(step, session.run(x), session.run(y), session.run(temperature), g))
        grad_norms.append(np.linalg.norm(g))

plt.plot(grad_norms)
```
